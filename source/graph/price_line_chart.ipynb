```python
import json
import matplotlib.pyplot as plt

def calculate_average(file_path):
    prices = []
    with open(file_path, 'r') as file:
        for line in file:
            json_object = json.loads(line)
            if 'price' in json_object:
                prices.append(json_object['price'])
    return sum(prices) / len(prices) if prices else 0

file_paths = ['part-00000-a71a515b-4323-472d-9ffe-8d32242e9eea-c000.json', 'part-00001-a71a515b-4323-472d-9ffe-8d32242e9eea-c000.json', 'part-00002-a71a515b-4323-472d-9ffe-8d32242e9eea-c000.json', 'part-00003-a71a515b-4323-472d-9ffe-8d32242e9eea-c000.json']

average_prices = [calculate_average(file_path) for file_path in file_paths]

quarters = ['Quarter 1', 'Quarter 2', 'Quarter 3', 'Quarter 4']
plt.figure(figsize=(10, 6))
plt.plot(quarters, average_prices, marker='o')
plt.title('Average Price per Quarter')
plt.xlabel('Quarter')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
```
